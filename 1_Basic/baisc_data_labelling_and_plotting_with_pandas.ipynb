# BASIC DATA LABELLING AND PLOTTING WITH PANDAS

```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\DELL\Desktop\70K Data Analysis\data_jobs.csv')
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

# Plotting with pandas

```python
job_counts.plot(kind = 'bar')
plt.title('Postings by Job Title')
plt.xlabel("")
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df[['job_posted_date','salary_year_avg']].dropna(subset=['salary_year_avg'])
```

```python
df.plot(x = 'job_posted_date', y = 'salary_year_avg', kind = 'line')
```

# Exercise

```python
# Calculating the median salaries based on the job title
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
```

```python
job_salary.plot(kind = 'barh')
plt.xlabel('Salary (USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')
plt.show()
```
